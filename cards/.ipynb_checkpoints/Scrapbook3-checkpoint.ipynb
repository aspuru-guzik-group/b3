```python
from rdkit.Chem import Draw
from rdkit.Chem.rdChemReactions import ReactionFromSmarts
from IPython.display import HTML
import re
tosmarts = lambda smi: Chem.MolToSmarts(getmol(smi))
def addidx(smiles, idx):
    mol = Chem.MolFromSmiles(smiles)
    for a in mol.GetAtoms():
        a.SetProp('molAtomMapNumber', str(idx))
    return Chem.MolToSmiles(mol)
def repairidx(mol):
    for a in mol.GetAtoms():
        props = a.GetPropsAsDict()
        if props.get('old_mapno'):
            a.SetProp('molAtomMapNumber', a.GetProp('old_mapno'))
    return mol
```

```python
trusted_vendors = ['Sigma-Aldrich', 'Combi-Blocks', 'Oakwood Products', 'Enamine', 
                   'Alfa Chemistry', 'Luminescence Technology Corp. (Lumtec)']
b_ends = ut.smiles_from_file('data/frags_boron.txt')

b_ends = [
    'C=CB1Oc2c(O1)cccc2',
    'CB(O)O',
    'CN1CC(=O)OB(OC(=O)C1)C',
    'O=C1C[N+]2([B-](O1)(C)OC(=O)C2)C',
    'CB1OC(C(O1)(C)C)(C)C',
    'CB1Nc2cccc3c2c(N1)ccc3',
]
def is_bc_bond(b):
    a1 = b.GetBeginAtom().GetSymbol()
    a2 = b.GetEndAtom().GetSymbol()
    if a1+a2 in ['BC', 'CB']:
        return True
    else:
        return False
b_frags = chopper.list_fragmentation(b_ends, cut_rules = [is_bc_bond], handle = '[H]')
b_frags = list(filter(lambda b: 'B' in b, b_frags))

def x_end(s):
    s = ut.replace_pattern(s, 'Br', '[Xe]', replaceAll=True)[0]
    s = ut.replace_pattern(s, 'I', '[Xe]', replaceAll=True)[0]
    return s
def b_end(s):
    def sub_b(sub):
        original = sub
        for b_frag in b_frags:
            sub = ut.replace_pattern(sub, b_frag, '[Be]', replaceAll=True)[0]
            if getmol(sub) is None:
                return original
        if '.' in sub:
            return original
        return sub
    return sub_b(s)
def get_vendor_prices(long_url):
    data = search.get_price(long_url)
    price_list = []
    for k, v in data.items():
        price_dict = {}
        if k == 'sigmaaldrich':
            price_dict['name'] = 'Sigma'
        elif k == 'oakwoodchemical':
            price_dict['name'] = 'Oakwood'
        elif k == 'enaminestore':
            price_dict['name'] = 'Enamine'
            
        price_dict['price'] = round(v['price'],2) if v['price'] < 1e7 else 'Nan'
        price_dict['availability'] = v['availability'] if len(v['availability']) > 0 else 'Nan'
        price_dict['url'] = v['url']
        price_list.append(price_dict)
    return price_list

def add_block_by_cid(cid, project):
    cid = str(cid)
    smiles = search.get_smiles_from_cids([cid])[0]
    block = b_end(x_end(smiles))
    prop = 'F'
    
    tag = ''
    for flag in ['Be', 'Ba', 'Bh', 'Xe']:
        if flag in block:
            tag += flag
    if tag == '':
        tag = 'None'
        
    urls = search.get_vendors_from_cid(cid, trusted_vendors)
    prices = get_vendor_prices(urls)
    
    details={
        'block': block,
        'smiles': smiles,
        'cid': cid, 
        'url': urls, 
        "prop": prop,
        "prices": prices
    }
    
    a2g2.put_smiles(
        block+'.'+smiles, 
        project = project,
        tags = [tag],
        details = details,
    )
    return

def get_info_by_smiles(smiles):
    cid = search.get_cid_by_smiles(smiles)
    if cid:
        cid = str(cid)
    else:
        return None, None, None, [None]
    block = b_end(x_end(smiles))
    
    tag = ''
    for flag in ['Be', 'Ba', 'Xe']:
        if flag in block:
            tag += flag
    if tag == '':
        tag = 'None'

    urls = search.get_vendors_from_cid(cid, trusted_vendors)
    return smiles, cid, urls, [tag]

```

```python
def min_price(a, b):
    ap = 1e20 if a['price'] == 'Nan' else a['price']
    bp = 1e20 if b['price'] == 'Nan' else b['price']
    if ap < bp:
        return a
    else:
        return b

def update_prices(mol, new_prices):
    for i, price in enumerate(mol.details['prices']):
        mol.details['prices'][i] = min_price(price, new_prices[i])

def add_block_by_smiles(smiles, project):
    try:
        smiles = getsmiles(getmol(smiles))
    except:
        return None
    
    # check if molecules already exist
    db_mols = a2g2.filter_mols(project)
    block = getsmiles(getmol(b_end(x_end(smiles))))
    match_mol = None
    for mol in db_mols:
        if mol.details['block'] == block:
            match_mol = mol
            break
            
    _, cid, urls, tags = get_info_by_smiles(smiles)
    if match_mol is not None:
        if cid not in match_mol.details['cid']:
            prices = get_vendor_prices(urls)
            match_mol.details['smiles'] += ('.'+smiles)
            match_mol.smiles += ('.'+smiles)
            match_mol.details['cid'] += ('!!'+cid)
            match_mol.details['url'] += ('!!'+urls)
            update_prices(match_mol, prices)
            match_mol.save()
            # remove svg
        return match_mol.inchikey
            
    else:
        prices = get_vendor_prices(urls)
        details={
            'block': block,
            'smiles': smiles,
            'cid': cid, 
            'url': urls, 
            'prices': prices,
            "prop": 'U',
        }
        new_mol = a2g2.put_smiles(
            block+'.'+smiles, 
            project = project,
            tags = tags,
            details = details,
            return_mol=True,
        )
        return new_mol['mol'].inchikey


print(add_block_by_smiles('CC(=O)C1=CC=C(S1)I', 'test'))
print(add_block_by_smiles('CC(=O)C1=CC=C(S1)Br', 'test'))
```
